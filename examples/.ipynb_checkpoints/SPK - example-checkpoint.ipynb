```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import pyspk as spk
```

```python
z = 0.125
fb_a = 8.44e-05
fb_pow = 0.275
k, sup = spk.sup_model(SO=200, z=z, fb_a=fb_a, fb_pow=fb_pow, k_max=10)
```

```python
plt.figure()
plt.plot(k, sup)
plt.xscale('log')

plt.axhline(1, c='k')
plt.xlabel('$k$ [$h$ Mpc$^{-1}$]');
plt.ylabel('$P_\mathrm{hydro}(k)/P_\mathrm{DM}(k)$');
```

```python
om = spk.best_mass_funct(200, .125, k)
```

```python
plt.figure()
plt.plot(k, om)
plt.xscale('log')
plt.yscale('log')

plt.xlabel('$k$ [$h$ Mpc$^{-1}$]');
plt.ylabel('$\\hat{M}_{k,200\mathrm{c}}\,\, [\mathrm{M}_\odot]$');
```
